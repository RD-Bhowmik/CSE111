```python
#Task-01
class Student:
    ID=0
    def __init__(self, name, department, age, cgpa):
        self.name=name
        self.department=department
        self.age=age
        self.cgpa=cgpa
        Student.ID+=1
        self.ID=Student.ID
    def get_details(self):
        print('ID:',Student.ID)
        print('Name:',self.name)
        print('Department:',self.department)
        print('Age:',self.age)
        print('CGPA:',self.cgpa)
    @classmethod
    def from_String(cls,info):
        name,department,age,cgpa=info.split('-')
        obj=cls(name,department,age,cgpa)
        return obj
s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details() 
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details() 
```

```python
#Task-02
class Assassin:    
    totalAssasinCount = 0    
    def __init__(self, name, successRate):        
        self.name = name
        self.successRate = successRate
        Assassin.totalAssasinCount += 1    
    @classmethod    
    def failureRate(cls, name, failureCount):
        return cls(name, 100 - failureCount)
        Assassin.totalAssasinCount += 1
    @classmethod
    def failurePercentage(cls, name, failurePercentage):
        return cls(name, 100 - failurePercentage)
        Assassin.totalAssasinCount += 1
    def printDetails(self):
        print("Name: {}".format(self.name))
        print("Success rate: {}%".format(self.successRate))
        print("Total number of Assassin: {}".format(self.totalAssasinCount))       
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#Task-03
class Passenger:
    count = 0
    def __init__(self, name):
        self.name = name
        self.bag_weight = 0
        self.fare = 450
        Passenger.count += 1
    def set_bag_weight(self, new_weight):
        self.bag_weight = new_weight
        if self.bag_weight > 50:
            self.fare += 100
        elif self.bag_weight > 20:
            self.fare += 50
    def printDetail(self):
        print(f"Name: {self.name}\n\
Bus Fare: {self.fare} taka")
print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
#Task-05
from datetime import date
class Employee:    
    def __init__(self, name, workingPeriod):
        self.name = name
        self.workingPeriod = workingPeriod
    @classmethod    
    def employeeByJoiningYear(cls, name, workingPeriod):
        return cls(name, date.today().year - workingPeriod)
    @staticmethod
    def experienceCheck(workingPeriod, gender):
        expStatus = ''
        if workingPeriod < 3:
            expStatus = " is not experienced"
        else:
            expStatus = " is experienced"
        if gender.upper() == "MALE":
            return "He" + expStatus
        else:
            return "She" + expStatus
employee1 = Employee('Dororo', 3)    
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, 'male'))
print(Employee.experienceCheck(3, 'female'))
```

```python
#Task-08
import math

class Cylinder:
    radius = 5
    height = 18
    def __init__(self, radius, height):
        print("Default radius=%.2f and height=%.2f" %(Cylinder.radius, Cylinder.height))
        Cylinder.radius = radius
        Cylinder.height = height
        print("Updated: radius=%.2f and height=%.2f" %(Cylinder.radius, Cylinder.height))
    @staticmethod
    def swap(height, radius):
        return Cylinder(radius, height)
    @staticmethod
    def changeFormat(string):
        values = string.split("-")
        radius = float(values[0])
        height = float(values[1])
        return Cylinder(radius, height)
    @staticmethod
    def area(radius, height):
        print("Area: %f" %(2 * math.pi * radius * (radius + height)))

    @staticmethod
    def volume(radius, height):
        print("Volume: %f" %(math.pi * radius * radius * height))

if __name__ == "__main__":
    c1 = Cylinder(0, 0)
    Cylinder.area(c1.radius, c1.height)
    Cylinder.volume(c1.radius, c1.height)
    print("=================================")
    c2 = Cylinder.swap(8, 3)
    Cylinder.area(c2.radius, c2.height)
    Cylinder.volume(c2.radius, c2.height)
    print("=================================")
    c3 = Cylinder.changeFormat("7-13")
    Cylinder.area(c3.radius, c3.height)
    Cylinder.volume(c3.radius, c3.height)
    print("=================================")
    Cylinder(0.3, 5.56).area(Cylinder.radius, Cylinder.height)
    print("=================================")
    Cylinder(3, 5).volume(Cylinder.radius, Cylinder.height)
  
```
